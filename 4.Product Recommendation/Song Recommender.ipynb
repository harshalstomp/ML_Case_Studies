```python
import turicreate
```

```python
songs = turicreate.SFrame("songs.SFrame")
```

```python
songs.print_rows(num_rows=1000,num_columns=6)
```

```python
eminem = songs[songs["artist"] == "Taylor Swift"]
eminem
```

```python
songs['song'].show()
```

```python
users = songs["user_id"].unique()
```

```python
len(users)
```

```python
train,test = songs.random_split(.8,seed=0)
```

```python
popu = turicreate.popularity_recommender.create(train,user_id="user_id",item_id="song")
```

```python
popu.recommend(users=[users[0]])
```

```python
users
```

```python
popu1 = turicreate.item_similarity_recommender.create(train,user_id="user_id",item_id="song")
```

```python
popu1.recommend(users=[users[0]])
```

```python
popu1.recommend(users=[users[1]])
```
