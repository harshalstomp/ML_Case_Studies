```python
adv_ft=[
'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'zipcode','condition','grade','waterfront','view',
'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'lat', 'long', 'sqft_living15','sqft_lot15', ]
```

```python
training_set, test_set = sale.random_split(.8,seed=0)

training_set.show()
```

```python
test_set.show()
```

```python
adv_ft_model = turicreate.linear_regression.create(training_set,target='price',features=adv_ft)
```

```python
print(adv_ft_model.evaluate(test_set))
```
